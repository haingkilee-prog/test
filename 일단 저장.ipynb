```python
baseball = 1234
s = 0
b = 0
answer = input("네자리 숫자를 입력하세요")
for i in range(4):
    if answer[i] == str(baseball)[i]:
        s += 1
    elif answer[i] in str(baseball):
        b += 1
    else:
        continue
print(f"스트라이크 : {s} \n볼 :{b}")
```
